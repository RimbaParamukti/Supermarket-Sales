```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/supermarket_sales - Sheet1.csv')
df.info()
```

**Data Cleaning**

```python
# Convert 'Date' to datetime
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')

# Convert 'Time' to datetime
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M')

df.info()
df
```

**EXPORT DATA FOR TABLEAU**

```python
# prompt: export and save to csv

df.to_csv('cleaned_supermarket_sales.csv', index=False)
```

**EXPLORING DATA**

**Total Sales by Branch**

```python
sum_sales_by_branch = df.groupby('Branch').sum(['Total']).round(2)
sum_sales_by_branch
```

**Total Sales by City**

```python
sum_sales_by_city = df.groupby('City')['Total'].sum().round(2)
sum_sales_by_city

# Yagon City only has branch A
# Mandalay City only has branch B
# Naypyitaw City only has branch C
```

**Total Sales by Product**

```python
sum_product = df.groupby('Product line').sum(['Total']).round(2)
sum_product
```

**Average Sales per Transaction by Customer Type**

```python
average_sales_customer_type = df.groupby('Customer type').sum(['Total']).round(2)
average_sales_customer_type
```

**Payment Method Distribution**

```python
payment_method = df['Payment'].value_counts()
payment_method
```

**Sales by Month**

```python
monthly = df['Date'].dt.to_period('M')
monthly_sales = df.groupby(monthly).sum(['Total']).round(2)
monthly_sales
```
